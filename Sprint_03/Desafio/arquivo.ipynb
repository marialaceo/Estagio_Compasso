### 1 - Lendo arquivo e retirando as duplicatas

* ```import pandas as pd```: Esta linha importa a biblioteca pandas.
* ```import matplotlib.pyplot as plt```: Esta linha importa a biblioteca Matplotlib, especificamente o módulo pyplot.
* ```df = pd.read_csv('googleplaystore.csv')```: Esta linha lê o arquivo CSV chamado ```'googleplaystore.csv'``` e o carrega em um DataFrame do pandas chamado df. 
* ```df.drop_duplicates(inplace=True)```: Esta linha remove quaisquer linhas duplicadas do DataFrame df. O parâmetro ```inplace=True``` garante que a remoção de duplicatas seja feita diretamente no DataFrame df, sem a necessidade de criar uma nova variável.
* ```df_clean = df.dropna()```: Esta linha remove todas as linhas que contêm valores ausentes ```(NaN)``` do DataFrame df e armazena o resultado em um novo DataFrame chamado ```df_clean```. Isso garante que o DataFrame df_clean contenha apenas linhas com dados completos.
* ```display(df_clean)```: Esta linha exibe o DataFrame df_clean, mostrando as primeiras e últimas linhas da tabela, assim como a estrutura geral dos dados limpos.

````python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('googleplaystore.csv')

df.drop_duplicates(inplace=True)
df_clean = df.dropna()
display(df_clean)
````

### 2 - Gráfico de barras - Top 5 apps por número de instalação

* ```top_5 = df_clean[['App', 'Installs']].groupby('App').sum().reset_index().head(5)```: Esta linha faz o seguinte:

   * Seleciona as colunas ```'App'``` e ```'Installs'``` do DataFrame ```df_clean```.
   * Agrupa os dados pelo nome do aplicativo ```('App')``` e soma as instalações ```('Installs')``` para cada grupo.
   * Reseta o índice do DataFrame resultante para que ```'App'``` volte a ser uma coluna regular.
   * Seleciona as primeiras 5 linhas do DataFrame resultante.

* ```top_5 = top_5.sort_values(by='Installs', ascending=False)```: Esta linha ordena o DataFrame ```top_5``` pelo número de instalações ```('Installs')``` em ordem decrescente, garantindo que os aplicativos com mais instalações apareçam primeiro.
* ```top_5_install = top_5['Installs']```: Esta linha seleciona a coluna ```'Installs'``` do DataFrame ```top_5``` e a armazena na variável ```top_5_install```.
* ```top_5_app = top_5['App']```: Esta linha seleciona a coluna ```'App'``` do DataFrame ```top_5``` e a armazena na variável ```top_5_app```.
* ```figura= plt.figure(figsize=(40,4))```: Esta linha cria uma nova figura para o gráfico com um tamanho específico (40 de largura e 4 de altura).
* ```figura.add_subplot(132)```: Esta linha adiciona um subplot (um gráfico dentro da figura) à figura criada. O parâmetro 132 significa que estamos criando um gráfico em uma grade de 1 linha por 3 colunas e que este é o segundo subplot da grade.
  
* ```plt.bar(top_5_app, top_5_install,color='skyblue',width=0.5)```: Esta linha cria um gráfico de barras com:

    * ```top_5_app``` nos rótulos do eixo ```x``` (nomes dos aplicativos).
    * ```top_5_install``` nos valores do eixo ```y ```(número de instalações).
    * A cor das barras definida como ```'skyblue'```.
    * A largura das barras definida como 0.5.
      
* ```plt.xlabel('App')```: Esta linha define o rótulo do eixo ```x``` como ```'App'```.
* ```plt.ylabel('Installs')```: Esta linha define o rótulo do eixo ```y``` como ```'Installs'```.
* ```plt.title('Top 5 Apps por Instalações')```: Esta linha define o título do gráfico como **'Top 5 Apps por Instalações'**.
* ```plt.show(figura)```: Esta linha exibe a figura e o gráfico criado.

````python
top_5 = df_clean[['App', 'Installs']].groupby('App').sum().reset_index().head(5)

top_5 = top_5.sort_values(by='Installs', ascending=False)

top_5_install = top_5['Installs']
top_5_app = top_5['App']

figura= plt.figure(figsize=(40,4))
figura.add_subplot(132)
plt.bar(top_5_app, top_5_install,color='skyblue',width=0.5)
plt.xlabel('App')
plt.ylabel('Installs')
plt.title('Top 5 Apps por Instalações')
plt.xticks()
plt.show(figura)
````

### 3 - Gráfico de pizza - Categorias de apps existentes no dataset de acordo com a frequência em que aparecem

* ```category_counts = df_clean[['Category']].groupby('Category').size().reset_index(name='Count')```:

   * ```df_clean[['Category']]```: Seleciona a coluna ```'Category'``` do DataFrame ```df_clean```.
   * ```groupby('Category')```: Agrupa os dados pela coluna ```'Category'```.
   * ```size()```: Conta o número de ocorrências em cada grupo, resultando no número de aplicativos em cada categoria.
   * ```reset_index(name='Count')```: Reseta o índice do DataFrame resultante e nomeia a coluna de contagem como ```'Count'```. Isso resulta em um DataFrame com duas colunas: ```'Category'``` e ```'Count'```, onde ```'Category'``` é a categoria do aplicativo e ```'Count'``` é o número de aplicativos nessa categoria.

* ```labels = category_counts['Category']```: Esta linha seleciona a coluna ```'Category'``` do DataFrame ```category_counts``` e a armazena na variável ```labels```.
* ```sizes = category_counts['Count']```: Esta linha seleciona a coluna ```'Count'``` do DataFrame ```category_counts``` e a armazena na variável sizes. Esta variável contém os tamanhos das fatias do gráfico de pizza, representando o número de aplicativos em cada categoria.
* ```figura = plt.figure(figsize=(6, 6))```: Esta linha cria uma nova figura para o gráfico com um tamanho específico de 6 polegadas por 6 polegadas.
  
* ```plt.pie(sizes, labels=labels, autopct=lambda p: f'{int(p * sum(sizes) / 100)}', startangle=140, colors=plt.cm.Paired.colors)```:

   * ```sizes```: Define os tamanhos das fatias do gráfico de pizza.
   * ```labels=labels```: Define os rótulos das fatias com os valores da variável labels.
   * ```autopct=lambda p: f'{int(p * sum(sizes) / 100)}'```: Define o formato dos rótulos das porcentagens dentro das fatias. Usa uma função lambda para calcular e exibir o número absoluto de aplicativos em vez de uma porcentagem. p é a porcentagem da fatia, sum(sizes) é o total de aplicativos, e ```sum(sizes) / 100``` calcula o número absoluto.
   * ```startangle=140```: Define o ângulo inicial do gráfico de pizza para 140 graus, para melhor visualização.
   * ```colors=plt.cm.Paired.colors```: Define a paleta de cores usada para as fatias do gráfico, utilizando a colormap ```'Paired'``` do ```Matplotlib```.

* ```plt.axis('equal')```: Esta linha assegura que o gráfico de pizza seja desenhado como um círculo, garantindo que a relação entre largura e altura seja igual.
* ```plt.title('Distribuição das Categorias')```: Esta linha define o título do gráfico como 'Distribuição das Categorias'.
* ```plt.show(figura)```: Esta linha exibe a figura e o gráfico criado.
    

````python
category_counts = df_clean[['Category']].groupby('Category').value_counts().reset_index(name='Count')

labels = category_counts['Category']
sizes = df_clean[['Category']].value_counts()

figura = plt.figure(figsize=(12, 15))
plt.pie(silabels=labels, radius=2.8 , colors=plt.cm.Paired.colors)
plt.axis('equal')
plt.title('Distribuição das Categorias')
plt.show(figura)
````

### 4 - App mais caro do dataset

* ```max_index = df_clean['Price'].idxmax()```:

   * ```df_clean['Price']```: Seleciona a coluna ```'Price'``` do DataFrame ```df_clean```.
   * ```.idxmax()```: Retorna o índice da linha que contém o valor máximo na coluna ```'Price'```. Em outras palavras, esta linha encontra o índice da linha onde o preço do aplicativo é o mais alto.

* ```top = df_clean.loc[max_index, ['App', 'Price']]```:

   * ```df_clean.loc[max_index, ['App', 'Price']]```: Utiliza o método ```.loc[]``` para acessar uma linha específica do DataFrame ```df_clean``` com base no índice ```max_index```.
   * ```max_index``` é o índice da linha que contém o preço máximo, conforme obtido na linha anterior.
   * ```['App', 'Price']``` são as colunas que queremos selecionar da linha especificada por ```max_index```. Então, esta linha de código seleciona os valores das colunas ```'App'``` e ```'Price'``` da linha onde o preço é o mais alto.

* ```display(top)```: Esta linha exibe o resultado armazenado na variável top.

````python
max_index = df_clean['Price'].idxmax()
top = df_clean.loc[max_index, ['App', 'Price']]
display(top)
````

### 6 - Top 10 apps por número de reviews

* ```.groupby('App')```: Agrupa os dados pela coluna ```'App'```.

* ```.reset_index()```: Reseta o índice do DataFrame resultante para que ```'App'``` volte a ser uma coluna regular.

### 7 - Cálculos do dataset (lista e valor)

#### Lista

* ```top_filmes_recentes = df_clean[['App','Last Updated']].groupby('App').max().reset_index().nlargest(5, 'Last Updated')```:

   * ```df_clean[['App', 'Last Updated']]```: Seleciona as colunas ```'App'``` e ```'Last Updated'``` do DataFrame ```df_clean```. Isso cria um novo DataFrame contendo apenas essas duas colunas.
   * ```.groupby('App')```: Agrupa o DataFrame resultante pela coluna ```'App'```. Isso significa que todos os registros com o mesmo nome de aplicativo serão agrupados juntos.
   * ```.max()```: Dentro de cada grupo, seleciona o valor máximo da coluna ```'Last Updated'```. Isso resulta no registro mais recentemente atualizado para cada aplicativo.
   * ```.reset_index()```: Reseta o índice do DataFrame resultante para que ```'App'``` volte a ser uma coluna regular, em vez de permanecer como índice.
   * ```.nlargest(5, 'Last Updated')```: Seleciona as 5 linhas com os maiores valores na coluna ```'Last Updated'```. Isso retorna os 5 aplicativos mais recentemente atualizados.
* ```top_filmes_recentes = top_filmes_recentes.sort_values(by='Last Updated', ascending=False)```:

   * ```top_filmes_recentes.sort_values(by='Last Updated', ascending=False)```: Ordena o DataFrame top_filmes_recentes pelos valores na coluna ```'Last Updated'``` em ordem decrescente. Isso garante que o aplicativo mais recentemente atualizado apareça primeiro na lista.

#### Valor

* ```classificacao_category = (df_clean['Category'] == 'EDUCATION').sum()```:

   * ```df_clean['Category']```: Seleciona a coluna ```'Category'``` do DataFrame ```df_clean```.
   * ```df_clean['Category'] == 'EDUCATION'```: Cria uma série de valores booleanos **(True ou False)** onde cada valor indica se a categoria do aplicativo é ```'EDUCATION'```. Para cada linha do DataFrame:
   * Se o valor na coluna ```'Category'``` for ```'EDUCATION'```, a expressão será ```True``` para essa linha.
   * Caso contrário, a expressão será ```False```.
   * ```.sum()```: Soma os valores booleanos. Somando os valores booleanos, você obtém o número de ocorrências onde a expressão é ```True```, ou seja, o número de aplicativos cuja categoria é ```'EDUCATION'```.


````python
top_filmes_recentes = df_clean[['App','Last Updated']].groupby('App').max().reset_index().nlargest(5, 'Last Updated')
top_filmes_recentes = top_filmes_recentes.sort_values(by='Last Updated', ascending=False)
display(top_filmes_recentes)
````

````python
classificacao_category = (df_clean['Category'] == 'EDUCATION').sum()
print(f'Existem {classificacao_category} apps categorizados como educativos')
````
